```python
import requests
import pandas as pd
```

```python

url = "https://api.finmindtrade.com/api/v4/data"
parameter = {
    "dataset": "TaiwanStockNews",
    "data_id":"2330",
    "start_date": "2023-01-01",
    "end_date": "2023-11-07",
}
data = requests.get(url, params=parameter)
data = data.json()
data = pd.DataFrame(data['data'])
```

```python
data.to_csv('data/news.csv', index=False)
print(data)
```
