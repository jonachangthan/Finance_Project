```python
import pandas as pd
import os
import glob
path = "data/stocks/"
```

```python
stocks = glob.glob(os.path.join(path, "*.csv"))
for stock in stocks:
    try:
        df = pd.read_csv(stock)
        df['DailyReturn'] = (df['close'] - df['close'].shift(1)) / df['close'].shift(1)
        df.to_csv(stock, index=False)
        print(stock)
    except:
        print("Error: ", stock)
```
